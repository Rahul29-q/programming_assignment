```python
# Program to Add Two Matrices
#Entering the first matrix
R=int(input("Enter row size:"))
C=int(input("Enter column size:"))
print("Enter the first matrix row wise:")
matrix1=[]
for i in range(0,R):
            a=[]
            for j in range(0,C):
                a.append(int(input()))
            matrix1.append(a)
#Entering the second matrix
print("Enter the second matrix row wise:")
matrix2=[]
for i in range(0,R):
            a=[]
            for j in range(0,C):
                a.append(int(input()))
            matrix2.append(a)
#Adding the given matrices
addition=[]
for i in range(0,R):
    a=[]
    for j in range(0,C):
        a.append(matrix1[i][j]+matrix2[i][j])
    addition.append(a)

#Printing the resultant matrix after addition
print("The addition of the given matrices is")
for i in range(0,R):
    
        print(addition[i]," ")
        print("\n")
        
```

```python
#Program to Multiply Two Matrices
# Entering the first matrix
import sys
R=int(input("Enter the number of rows of the first matrix:"))
C=int(input("Enter the number of columns of the first matrix:"))
print("Enter the first matrix row wise:")
matrix1=[]
for i in range(0,R):
            a=[]
            for j in range(0,C):
                a.append(int(input()))
            matrix1.append(a)
#Entering the second matrix
r=int(input("Enter the number of rows of the second matrix:"))
c=int(input("Enter the number of columns of the second matrix:"))
if C!=r:
    print("Sorry! operation can not be performed due to the inappropriate size of the matrix")
    sys.exit()
else:
    print("Enter the second matrix row wise:")            
matrix2=[]
for i in range(0,r):
            a=[]
            for j in range(0,c):
                a.append(int(input()))
            matrix2.append(a)
#Initializing the matrix that stores the values of matrix multiplication 
multiplication=[]
for i in range(0,R):
    a=[]
    for j in range(0,c):
        a.append(0)
    multiplication.append(a)
#Computing the multiplication
for i in range(0,R):
    
    for j in range(0,c):
        for k in range(0,C):
                       
            multiplication[i][j]=matrix1[i][k]*matrix2[k][j]+multiplication[i][j]
#Printing the resultant matrix after multiplication
print("The multiplication of the given matrices is")
for i in range(0,R):
    
    
        print(multiplication[i]," ")
        
        
```

```python
#Program to Transpose a Matrix
# Entering the  matrix
R=int(input("Enter the number of rows of the matrix:"))
C=int(input("Enter the number of column of the matrix:"))
print("Enter matrix row wise:")
matrix1=[]
for i in range(0,R):
            a=[]
            for j in range(0,C):
                a.append(int(input()))
            matrix1.append(a)
#Initializing the matrix that stores the values of transposed matrix
transpose=[]
for i in range(0,C):
    a=[]
    for j in range(0,R):
        a.append(matrix1[j][i])
    transpose.append(a)
#Printing the transpose matrix 
print("The transpose of the given matrix is")
for i in range(0,C):
    
    
        print(transpose[i]," ")
```

```python
#Program to Sort Words in Alphabetic Order
# To take input from the user
my_str = input("Enter a string: ")

# breakdown the string into a list of words
words = [word.lower() for word in my_str.split()]

# sort the list
words.sort()

# display the sorted words

print("The sorted words are:")
for word in words:
   print(word)
```

```python
# define punctuation
punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''

my_str = "Hello!!!, he said ---and went."

# To take input from the user
my_str = input("Enter a string: ")

# remove punctuation from the string
no_punct = ""
for char in my_str:
   if char not in punctuations:
       no_punct = no_punct + char

# display the unpunctuated string
print(no_punct)
```
